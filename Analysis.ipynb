```python
def func(x):
    if pd.isna(x):
        return None
    if x == "-1":
        return x
    elif "-" in x:
        return x.strip().split("-")[0]
    else:
        return x

df["Experience Year"] = df['Experience Year'].apply(func)
```

```python
df['Experience Year'] = df['Experience Year'].astype(dtype='float')
```
