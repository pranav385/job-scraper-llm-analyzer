```python
def slow_type(element, text, delay_range=(0.05, 0.15)):
    for char in text:
        element.send_keys(char)
        time.sleep(random.uniform(*delay_range))
```

```python
def human_scrolls(browser, max_scrolls=10):
    last_height = browser.execute_script("return document.body.scrollHeight")

    for i in range(max_scrolls):
        scroll_distance = random.randint(300, 800)

        browser.execute_script(f"window.scrollBy(0, {scroll_distance});")

        time.sleep(random.uniform(0.8, 2.5))

        if random.random() < 0.2:
            time.sleep(random.uniform(3, 5))

        new_height = browser.execute_script("return window.scrollY + window.innerHeight")
        total_height = browser.execute_script("return document.body.scrollHeight")

        if new_height >= total_height - 50:
            break
```

```python
def human_like_scroll(browser, scrollable_div, min_scrolls=6, max_scrolls=12):
    last_scroll = -1
    same_scroll_count = 0
    
    for i in range(random.randint(min_scrolls, max_scrolls)):
        scroll_increment = random.randint(100, 400)

        if random.random() < 0.2:
            scroll_increment = random.randint(600, 1200)

        browser.execute_script(
            "arguments[0].scrollTop += arguments[1];", 
            scrollable_div, scroll_increment
        )

        if random.random() < 0.3:
            time.sleep(random.uniform(2.0, 4.0))
        else:
            time.sleep(random.uniform(0.8, 1.8))

        current_scroll = browser.execute_script("return arguments[0].scrollTop;", scrollable_div)
        
        if current_scroll == last_scroll:
            same_scroll_count += 1
            if same_scroll_count > 2:
                break
        else:
            same_scroll_count = 0
        
        last_scroll = current_scroll

        if i % 4 == 0:
            time.sleep(random.uniform(3.0, 6.0))
```

```python
if random.random() < 0.7:
    time.sleep(random.uniform(5, 8))
else:
    time.sleep(random.uniform(8, 12))

human_scrolls(browser, max_scrolls=6)


job_title = input("Enter job title (e.g. Data Analyst, Java Developer): ").strip()
hours = int(input("Enter time filter in hours (e.g. 24, 72, 168): ").strip())

time_in_seconds = hours * 3600  

location_geoId = "102713980"  

job_title_encoded = quote(job_title)

# job_url = f"https://www.linkedin.com/jobs/search/?f_E=1%2C2&f_TPR=r{time_in_seconds}&geoId={location_geoId}&keywords={job_title_encoded}"

job_url = f"https://www.linkedin.com/jobs/search/?f_E=1%2C2%2C3&f_TPR=r{time_in_seconds}&geoId={location_geoId}&keywords={job_title_encoded}"

# job_url = "https://www.linkedin.com/jobs/search/?f_E=1%2C2%2C3&f_TPR=r86400&geoId=102713980&keywords=Data%20Analyst"

print("Generated Job URL:", job_url)
```

```python
browser.get(job_url)


if random.random() < 0.7:
    time.sleep(random.uniform(6, 12))
else:
    time.sleep(random.uniform(20, 40))
```

```python
all_job_list = []
session_job_limit = 500  
link_set = set()
cnt = 0

while True:

    # parent_div = WebDriverWait(browser, 10).until(
    #     EC.presence_of_element_located((By.CSS_SELECTOR, "div.scaffold-layout__list"))
    # )

    parent_div = WebDriverWait(browser, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.scaffold-layout__list > div"))
    )
    
    # child_divs = parent_div.find_elements(By.TAG_NAME, "div")
    # scrollable_div = None
    scrollable_div = parent_div

    if not scrollable_div:
        print("No scrollable div found, stopping.")
        break

    last_scroll = 0

    for i in range(random.randint(4, 8)):
        scroll_increment = random.randint(300, 800)
        browser.execute_script(
            "arguments[0].scrollTop += arguments[1];", scrollable_div, scroll_increment
        )
        
        time.sleep(random.uniform(1.0, 2.0))
    
        current_scroll = browser.execute_script("return arguments[0].scrollTop;", scrollable_div)
        
        if current_scroll == last_scroll:
            break
        last_scroll = current_scroll
    
        if i % 3 == 0:
            time.sleep(random.uniform(5, 8))

    WebDriverWait(browser, 5).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "li.ember-view"))
    )
    all_jobs = scrollable_div.find_elements(By.CSS_SELECTOR, "li.ember-view")

    if random.random() < 0.3:
        all_jobs.reverse()

    for idx, job_card in enumerate(all_jobs):

        job_name = job_link = company_name = location = job_description = None
        
        if len(all_job_list) >= session_job_limit:
            print("Reached session job limit. Stopping for safety.")
            break

        try:
            ActionChains(browser).move_to_element(job_card).perform()
            time.sleep(random.uniform(0.5, 1.5))

            browser.execute_script("arguments[0].scrollIntoView(true);", job_card)
            time.sleep(random.uniform(0.8, 2.0))
            job_card.click()
            time.sleep(random.uniform(2.5, 5.0))

            WebDriverWait(browser, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "div.jobs-search__job-details--wrapper"))
            )

            # --------------------
            # Job info extraction
            # --------------------
            try:
                job_container = job_card.find_element(By.CLASS_NAME, "job-card-list__entity-lockup")
                content_div = job_container.find_element(By.CLASS_NAME, "artdeco-entity-lockup__content")

                try:
                    job_title_elem = content_div.find_element(By.CLASS_NAME, "artdeco-entity-lockup__title")
                    job_name = job_title_elem.find_element(By.TAG_NAME, "strong").text.strip()
                    job_link = job_title_elem.find_element(By.TAG_NAME, "a").get_attribute("href").strip()
                except Exception as e:
                    print(f"‚ö†Ô∏è Could not extract title/link: {e}")
                    job_name, job_link = None, None

                # company
                try:
                    company_elem = content_div.find_element(By.CLASS_NAME, "artdeco-entity-lockup__subtitle")
                    company_name = company_elem.text.strip()
                except Exception as e:
                    print(f"‚ö†Ô∏è Could not extract company: {e}")
                    company_name = None

                # location
                try:
                    location_elem = content_div.find_element(By.CLASS_NAME, "artdeco-entity-lockup__caption")
                    location = location_elem.text.strip()
                except Exception as e:
                    print(f"‚ö†Ô∏è Could not extract location: {e}")
                    location = None

                # avoid duplicates
                if job_link and job_link in link_set:
                    print('Already Scrapped')
                    continue
                if job_link:
                    link_set.add(job_link)

            except Exception as e:
                print("job_card", job_card)
                print(f"Error extracting job container: {e}")

            # --------------------
            # Job description
            # --------------------
            try:
                job_desc_scroll = WebDriverWait(browser, 5).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, "div.jobs-search__job-details--wrapper"))
                )

                last_height = 0
                for _ in range(random.randint(5, 12)):
                    increment = random.randint(200, 600)
                    browser.execute_script("arguments[0].scrollTop += arguments[1];", job_desc_scroll, increment)
                    time.sleep(random.uniform(1.0, 2.0))
            
                    new_height = browser.execute_script("return arguments[0].scrollTop;", job_desc_scroll)
                    if new_height == last_height:
                        break
                    last_height = new_height

                try:
                    job_desc_div = WebDriverWait(browser, 5).until(
                        EC.presence_of_element_located((By.ID, "job-details"))
                    )
                    job_description = job_desc_div.text.strip()
                except Exception as e:
                    print(f"‚ö†Ô∏è Could not extract job description: {e}")
                    job_description = None

            except Exception as e:
                print(f"Description scroll failed: {e}")
                job_description = None

            # print(job_description)
            print("*"*100)

            print(f"{idx} | Title: {job_name} | Company: {company_name} | Location: {location} | "
              f"Link length: {len(job_link) if job_link else 0} | "
              f"JD length: {len(job_description) if job_description else 0}")
            
            print("*"*100)
            
            all_job_list.append([job_name, job_link, company_name, location, job_description])

        except Exception as e:
            print(f"Skipping job at index {idx}: {e}")
            continue

    # --------------------
    # Next page navigation
    # --------------------
    next_buttons = browser.find_elements(By.CSS_SELECTOR, "button.jobs-search-pagination__button--next")

    if next_buttons:
        next_button = next_buttons[-1]
    
        try:
            browser.execute_script("arguments[0].scrollIntoView({block: 'center'});", next_button)
            time.sleep(random.uniform(1, 2))
    
            ActionChains(browser).move_to_element(next_button).perform()
            time.sleep(random.uniform(0.8, 2.0))
    
            if random.random() < 0.1:
                time.sleep(random.uniform(2, 4))
    
            next_button.click()
    
            WebDriverWait(browser, 12).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "div.scaffold-layout__list"))
            )
            time.sleep(random.uniform(3, 6))
    
        except Exception as e:
            print(f"Failed to click next button: {e}")
            break
    else:
        print("No more next button, scraping finished.")
        break

print(f"Total jobs scraped: {len(all_job_list)}")
```

```python
import random, time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

max_attempts = 4
browser.get("https://www.linkedin.com/feed/")

for attempt in range(1, max_attempts + 1):
    try:
        print(f"üîÑ Attempt {attempt} to log out...")
        browser.refresh()
        
        # Open profile dropdown
        profile_dropdown = WebDriverWait(browser, 10).until(
            EC.element_to_be_clickable((By.CLASS_NAME, "global-nav__primary-link-me-menu-trigger"))
        )
        profile_dropdown.click()
        time.sleep(2 + random.uniform(0.5, 1.5))

        # Try both possible sign out locators
        try:
            # Case 1: Normal link
            sign_out_btn = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.LINK_TEXT, "Sign Out"))
            )
        except:
            # Case 2: Button with span text
            sign_out_btn = WebDriverWait(browser, 5).until(
                EC.element_to_be_clickable((By.XPATH, "//span[text()='Sign out' or text()='Sign Out']/ancestor::button"))
            )

        sign_out_btn.click()
        time.sleep(3 + random.uniform(0.5, 1.5))

        print("‚úÖ Successfully logged out!")
        break  # Exit the loop if logout succeeds

    except Exception as e:
        print(f"‚ö†Ô∏è Logout attempt {attempt} failed: {e}")
        if attempt == max_attempts:
            print("‚ùå All attempts failed. Could not log out.")
```
